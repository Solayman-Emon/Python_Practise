```python
# create a dataframe

from numpy.random import randint 
arr = randint(0, 100, 16).reshape(4,4)
print(arr)
rows = "A B C D".split()
cols = "E F G F".split()
df = pd.DataFrame(arr, index = rows, columns = cols)
df
```

```python
# array selection & indexing 

from numpy.random import randint 
arr = randint(0, 100, 12).reshape(4, 3)
print(arr)
rows = "A B C D".split()
cols = "E F G".split()
df3 = pd.DataFrame(arr, index = rows, columns = cols)
print(df3)
# print(df3["B"])      # row wise indexing is not possible, only column wise possible
print(df3["E"])        # single column value
print(df3[["E", "F"]]) # multiple column, list of list

# row access

print(df3.loc["C"])    # C row value
print(df3.iloc[2])     # also possible access row by numeric values
```

```python
# add & delet column

df3["new"] = df3["E"] + df3["F"]
print (df3)
print(df3.drop("new", axis = 1))   # delet a column
print(df3)                         # but the new column doesn't delet permanently
df3["new"] = df3["E"] + df3["F"]   # add agian
print(df3)
print(df3.drop("new", axis = 1, inplace = True))  # delet permanently
print(df3)

# delet row

print(df3.drop("C", axis = 0, inplace = True))
print(df3)
```
